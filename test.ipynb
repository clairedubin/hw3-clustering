```python
import cluster
import math
import numpy as np
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_score
```

```python
from cluster import (
  make_clusters, 
  plot_clusters, 
  plot_multipanel)
```

```python
# here we are directly controlling the dimensionality of our data 
#   1000 observations 
#   200 features 
#   3 clusters)
d_clusters, d_labels = make_clusters(n=1000, m=200, k=3)

kmeans = cluster.KMeans(k=3)
silhouette = cluster.Silhouette()
predicted_labels=kmeans.predict(d_clusters)
scores = silhouette.score(d_clusters, d_labels)

plot_multipanel(d_clusters, d_labels, predicted_labels, score=scores)
```

```python
# here we are making tightly clustered data by lowering the scale param
t_clusters, t_labels = make_clusters(scale=0.3) 

kmeans = cluster.KMeans(k=3)
silhouette = cluster.Silhouette()

scores = silhouette.score(t_clusters, t_labels)

plot_multipanel(t_clusters, t_labels, kmeans.predict(t_clusters), score=scores)
```

```python
# here we are making loosely clustered data by increasing the scale param
l_clusters, l_labels = make_clusters(scale=2) 

kmeans = cluster.KMeans(k=3)
silhouette = cluster.Silhouette()
predicted_labels = kmeans.predict(l_clusters)
scores = silhouette.score(l_clusters, l_labels)

plot_multipanel(l_clusters, l_labels, predicted_labels, score=scores)
```

```python
# here we are making many clusters by adjusting the `k` param
m_clusters, m_labels = make_clusters(k=10)

kmeans = cluster.KMeans(k=10)
silhouette = cluster.Silhouette()
predicted_labels = kmeans.predict(m_clusters)
scores = silhouette.score(m_clusters, m_labels)

plot_multipanel(m_clusters, m_labels, predicted_labels, score=scores)
```
